```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

This code works because;
  
1. input() asks the user to enter a number.

2. int() converts the entered value into a number.

3. The number is stored in the variable num.

4. for i in range(1, 11) runs a loop from 1 to 10.

5. Each time the loop runs, i gets a new value (1, 2, 3, …, 10).

6. num * i multiplies the entered number by the current value of i.

7. print() displays the multiplication result.

8. The loop repeats until it reaches 10.

9.  The program then ends.
